```python
#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv("pokemon.csv")
```

#EDA

```python
train.shape
```

```python
train
```

```python
train.info()
```

```python
train.isnull().sum()
```

```python
sns.heatmap(train.isnull(), yticklabels=False, cbar=False, cmap="viridis")
```

Defining feature and Label

```python
y = train["target"].values
y.shape
```

```python
X = train.drop(columns = ["target"], inplace = False)
X = X.values
X.shape
```

```python
plt.imshow(X[5].reshape(100,100,3))
```

```python
#Splitting Data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state = 0)
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

#Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
DTmodel = DecisionTreeClassifier()
```

```python
DTmodel.fit(X_train, y_train)
y_pred_DT = DTmodel.predict(X_test)
y_pred_DT
```

*Decision Tree Score*

```python
DTmodel.score(X_test, y_test)
```

*Prediction and Comparison*

```python
for i in [2, 6, 8, 10 , 23, 60, 56, 49]:
  random_img = X_test[i]
  plt.imshow(random_img.reshape(100,100,3))
  plt.show()
  print("Actual Label: ", y_test[i])
  p = DTmodel.predict([random_img])
  print("Predicted Label: ", p)
```

#KNN

```python
from sklearn.neighbors import KNeighborsClassifier
KNNmodel = KNeighborsClassifier(n_neighbors = 17)
```

```python
KNNmodel.fit(X_train, y_train)
y_pred_KNN = KNNmodel.predict(X_test)
y_pred_KNN
```

*KNN Score*

```python
KNNmodel.score(X_test, y_test)
```

*Prediction and Comparison*

```python
for i in [2, 6, 8, 10 , 23, 60, 56, 49]:
  random_img = X_test[i]
  plt.imshow(random_img.reshape(100,100,3))
  plt.show()
  print("Actual Label: ", y_test[i])
  p = KNNmodel.predict([random_img])
  print("Predicted Label: ", p)
```

#Logistic Regression

One vs all classification for multiclass problem


```python
from sklearn.linear_model import LogisticRegression
LRmodel = LogisticRegression()
```

```python
LRmodel.fit(X_train, y_train)
y_predict_LR = LRmodel.predict(X_test)
y_predict_LR
```

*Logistic Regression Score*

```python
LRmodel.score(X_test, y_test)
```

*Prediction and Comparison*

```python
for i in [2, 6, 8, 10 , 23, 60, 56, 49]:
  random_img = X_test[i]
  plt.imshow(random_img.reshape(100,100,3))
  plt.show()
  print("Actual Label: ", y_test[i])
  p = LRmodel.predict([random_img])
  print("Predicted Label: ", p)
```

#Conclusion


Decision Tree Score is the least (0.7868)  and Logistic Regression (0.9836) score is max
Thus, for given data Logistic Regression model is the best
